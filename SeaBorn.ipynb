# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
Five plots:
    * Line Plot
    * Scatterplot
    * Barplot
    * Boxplot
    * countplot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fmri_data = sns.load_dataset("fmri")
sns.set(style="whitegrid")  # Set the style of the plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal Over Time for Different Events and Regions")
plt.legend(title="Event", loc="upper right")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic_data = sns.load_dataset("titanic")
titanic_data
```

```python
sns.set(style="whitegrid")
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
sns.boxplot(x='pclass', y='age', data=titanic_data, ax=axes[0])
axes[0].set_title('Box Plot of Age by Pclass')
sns.boxplot(x='pclass', y='fare', data=titanic_data, ax=axes[1])
axes[1].set_title('Box Plot of Fare by Pclass')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
diamonds = sns.load_dataset("diamonds")
diamonds
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple="stack")
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Histogram of Diamond Prices by Cut")
plt.legend(title="Cut", loc="upper right")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris_data = sns.load_dataset("iris")
sns.set(style="ticks")
sns.pairplot(iris_data, hue="species", markers=["o", "s", "D"])
plt.show()
```
